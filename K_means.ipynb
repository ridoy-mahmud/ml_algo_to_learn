```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans, k_means
```

```python
ds = pd.read_csv("./datasets/Salary_Data.csv")
ds.info()
```

```python
ds.describe()
```

```python
X = ds[['YearsExperience', 'Salary']]
```

```python
 means = KMeans(n_clusters=3, random_state=42)
 means.fit(X)
 ds['Cluster'] = means.labels_
```

```python
plt.scatter(ds['YearsExperience'], ds['Salary'], c=ds['Cluster'], cmap='viridis')
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("K-Means Clustering (Experience vs Income)")
plt.show()
```
