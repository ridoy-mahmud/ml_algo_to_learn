```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("segmented_customers.csv")
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
X=df[["CustomerID","Age","Annual Income (k$)"]]
```

```python
kms = KMeans(n_clusters=3, random_state=42)
kms.fit(X)
df["cluster"] = kms.labels_
```

```python
plt.scatter(df["CustomerID"],df["Age"],df["Annual Income (k$)"],c=df["cluster"],cmap="viridis")
plt.title("K measn Clustering")
```
